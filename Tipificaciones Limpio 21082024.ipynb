```python
import pandas as pd
import re
from unidecode import unidecode

def clean_text(text):
    if pd.isna(text):  # Manejar valores nulos
        return ''
    
    text = str(text)  # Asegurarse de que 'text' es una cadena de texto
    text = text.lower()
    
    # Eliminar signos de puntuación y exclamación
    text = re.sub(r'[^\w\s]', '', text)  # Elimina todos los caracteres que no son palabras o espacios
    
    # Reemplazar múltiples espacios por un solo espacio
    text = re.sub(r'\s+', ' ', text).strip()
    
    # Normalizar acentos y caracteres especiales
    text = unidecode(text)

    # Reemplazos específicos
    replacements = [
        "cliente indica que ", "cliente dice que ", "cliente comenta que "
    ]
    
    for phrase in replacements:
        text = text.replace(phrase, "")
    return text

    
```

```python
import pandas as pd

import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter
import string
def analyze_text(text):
    # Convertir a minúsculas
    text = text.lower()

    text = text.replace("cliente indica que ","")
    text = text.replace("cliente satisfecho","")



    text = unidecode(text)

    # Tokenizar el texto
    words = word_tokenize(text)

    # Eliminar puntuación
    words = [word for word in words if word.isalpha()]

    # Cargar stopwords en español
    stop_words = set(stopwords.words('spanish'))

    # Eliminar stopwords
    filtered_words = [word for word in words if word not in stop_words]

    # Contar la frecuencia de las palabras
    word_freq = Counter(filtered_words)

    # Obtener las palabras más comunes
    most_common_words = word_freq.most_common()

    # Calcular bigramas
    bigrams = list(nltk.bigrams(filtered_words))

    # Contar la frecuencia de los bigramas
    bigram_freq = Counter(bigrams)

    # Obtener los bigramas más comunes
    most_common_bigrams = bigram_freq.most_common()

    # Calcular trigramas
    trigrams = list(nltk.trigrams(filtered_words))

    # Contar la frecuencia de los trigramas
    trigram_freq = Counter(trigrams)

    # Obtener los trigramas más comunes
    most_common_trigrams = trigram_freq.most_common()

    return most_common_words, most_common_bigrams, most_common_trigrams

text_column = 'Comentario'
full_text = ' '.join(no_identified_comments[text_column].astype(str).tolist())
common_words, common_bigrams, common_trigrams = analyze_text(full_text)
```

```python

def normalize_text(text):
    return unidecode(text.lower())

def normalizacion_textos(documents):
    new_corpus = []
    for doc in documents:
        cleaned_doc = re.sub(r'[^a-zA-Z0-9\s{1}áéíóúÁÉÍÓÚ]', '', doc).lower().strip().rstrip('\n').rstrip('\r\n')
        cleaned_doc = normalize_text(cleaned_doc)
        new_corpus.append(cleaned_doc)
    return new_corpus
#

def classify_comment(comment):
    # Normalizar el texto del comentario
    comment = normalize_text(comment)
    comment = normalizacion_textos([comment])[0]

    # Verificar si el comentario es vacío
    if not comment.strip():  # Usar strip() para eliminar espacios en blanco
        return 'vacío'
    
    # Verificar combinaciones de palabras clave
    for theme, keyword_combinations in themes.items():
        for combination in keyword_combinations:
            if all(keyword in comment for keyword in combination):
                return theme
    if 'satisfecho' in comment or 'recomendado' in comment or 'excelente' in comment or 'satisfaccion' in comment or 'bueno' in comment or 'todo bien' in comment or 'todo muy bien' in comment:
        return 'Buena experiencia'
    return 'Tema no identificado'
```
